```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return  True
```

```python
def convert_to_num(x):
    tokens = str(x).split('-')
    if len(tokens) == 2:
            return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df[~df.total_sqft.apply(is_float)]
```

```python
df.total_sqft=df.total_sqft.apply(convert_to_num)
```

```python
df['price_per_sqft'] = (df.total_sqft) * (df.price)
```
